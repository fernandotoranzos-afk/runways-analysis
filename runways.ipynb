## 📊 Carga de datos

Se carga el dataset de pistas de aeropuertos para analizar características como longitud, superficie y estado operativo.

Objetivo: comprender la estructura inicial del dataset y detectar posibles problemas de calidad.

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ferna\Downloads\python\data\raw\runways.csv")
df.head()
```

## 🧹 Limpieza de datos

Se estandarizan los nombres de columnas y valores categóricos para evitar inconsistencias en el análisis.

Insight:
Se detectaron variaciones en la columna "surface" (ej: TURF, Turf, turf), lo que puede generar duplicaciones en el análisis.

```python
df.columns = df.columns.str.strip().str.lower()
df["surface"] = df["surface"].str.lower().str.strip()
```

```python
df["length_norm"] = (
    df["length_ft"] - df["length_ft"].min()
) / (
    df["length_ft"].max() - df["length_ft"].min()
)
```

```python
df["length_norm"].describe()
```

```python
df.info()
df.describe()
df.columns
```

## 📏 Normalización de variables

Se normaliza la longitud de las pistas para llevar los valores a una escala entre 0 y 1.

Insight:
Permite comparar pistas de diferentes tamaños sin sesgo por magnitud.

```python
df["length_norm"] = (df["length_ft"] - df["length_ft"].min()) / (df["length_ft"].max() - df["length_ft"].min())
df[["length_ft", "length_norm"]].head(10)
```

## 🔢 Codificación de variables categóricas

Se convierte la variable "surface" a formato numérico.

Insight:
Esto permite utilizar la variable en modelos o análisis cuantitativos.

```python
df["surface_code"] = df["surface"].astype("category").cat.codes
df["surface"].unique()
df["surface_code"].unique()
df[["surface", "surface_code"]].head(10)
```

```python
df_dummy = pd.get_dummies(df, columns=["surface"])
df_dummy.head()
```

```python
df["area"] = df["length_ft"] * df["width_ft"]
df[["length_ft", "width_ft", "area"]].head(10)
```

## 📊 Análisis por tipo de superficie

Se agrupan las pistas según su superficie para analizar diferencias en longitud.

Insights:
- Las pistas pavimentadas presentan mayor longitud promedio.
- Las superficies no pavimentadas están asociadas a pistas más cortas.
- Existe una mayor cantidad de pistas en superficies simples (ej: turf).

```python
df.groupby("surface")["length_ft"].agg(["mean", "max", "min", "count"])
```

## 🔄 Análisis cruzado: superficie vs iluminación

Se analiza la relación entre la longitud de las pistas y si están iluminadas.

Insights:
- Las pistas iluminadas tienden a ser más largas.
- La iluminación está asociada a mayor infraestructura.

```python
pivot = df.pivot_table(
    values="length_ft",
    index="surface",
    columns="lighted",
    aggfunc="mean"
)

pivot
```

## 🧠 Creación de nuevas variables

Se crea la variable "area" para representar el tamaño real de la pista.

Insights:
- El área permite identificar aeropuertos de mayor escala.
- Se clasifica el tamaño en categorías (small, medium, large) para facilitar análisis.

```python
df["area"] = df["length_ft"] * df["width_ft"]
```

```python
def clasificar_pista(x):
    if x < 50000:
        return "small"
    elif x < 200000:
        return "medium"
    else:
        return "large"

df["size_category"] = df["area"].apply(clasificar_pista)
```

## 📦 Dataset final

Se genera un dataset limpio y optimizado para su uso en Power BI.

Columnas incluidas:
- length_ft
- width_ft
- length_norm
- surface_code
- lighted
- area
- size_category

Insight:
El dataset está listo para visualización y análisis en herramientas BI.

```python
df_final = df[[
    "length_ft",
    "width_ft",
    "length_norm",
    "surface_code",
    "lighted",
    "area",
    "size_category"
]]
```

```python
df_final.to_csv("runways_clean.csv", index=False)
```
